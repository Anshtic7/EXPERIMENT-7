```python
# Factorial of number
n = int(input("Enter a non-negative number: "))
if n < 0:
  print("Factorial is not defined for negative numbers.")
elif n == 0:
  print("The factorial of 0 is 1")
else:
  fact = 1
  i = 1
  while i <= n:
    fact *= i
    i += 1
  print(f"The factorial of {n} is {fact}")
```

```python
# Write a python program using while loop for fibonacci series.

no_of_terms = int(input("Enter the number of terms: "))

a, b = 0, 1
count = 0

if no_of_terms <= 0:
  print("Please enter a positive integer.")
elif no_of_terms == 1:
  print("Fibonacci sequence up to 1 term:")
  print(a)
else:
  print("Fibonacci sequence:")
  while count < no_of_terms:
    print(a)
    nth = a + b
    # Updated values
    a = b
    b = nth
    count += 1
```

```python
# Write a python program using while loop to print fibonacci series with a limiting number.
limit = int(input("Enter the limit: "))
a = 0
b = 1

while a <= limit:
  print(a, end = " ")
  a, b = b, a + b
```

```python
# Program to reverse a number using while loop

num = int(input("Enter a number: "))
rev = 0
while num > 0:
    digit = num % 10          # Get last digit
    rev = rev * 10 + digit
    num = num // 10           # Remove last digit
print("Reversed number is:", rev)
```

```python
# Write a python program using while loop to check palindrome number/string.

# --- Check Palindrome Number ---
print("\n--- Palindrome Number Check ---")
original_num = int(input("Enter an integer to check for palindrome: "))
num = original_num
reversed_num = 0

while num > 0:
  digit = num % 10
  reversed_num = reversed_num * 10 + digit
  num = num // 10

if original_num == reversed_num:
  print(f"{original_num} is a palindrome number.")
else:
  print(f"{original_num} is not a palindrome number.")

# --- Check Palindrome String ---
print("\n--- Palindrome String Check ---")
original_string = input("Enter a string to check for palindrome: ")
cleaned_string = "".join(char.lower() for char in original_string if char.isalnum())

reversed_string = ""
index = len(cleaned_string) - 1
while index >= 0:
  reversed_string += cleaned_string[index]
  index -= 1

if cleaned_string == reversed_string:
  print(f"'{original_string}' is a palindrome string.")
else:
  print(f"'{original_string}' is not a palindrome string.")
```

```python
# Another method for pallindrome number/string
st = input("Enter a string: ")
rev = st[::-1] #in-built function which itself creates a reverse string of orginal string
if st == rev:
    print(st, "is a pallindrome string")
else:
  print(st, "is not a pallindrome string")
```

```python
# Write a python program using while loop to count digits in a number.

num = int(input("Enter a number: "))
count = 0

# Handle the case of 0 separately as the loop condition num > 0 won't execute for it.
if num == 0:
  count = 1
else:
  # Make the number positive for counting digits correctly
  num = abs(num)
  while num > 0:
    num = num // 10
    count += 1

print(f"The number of digits is: {count}")
```

# **break**

The break statement in Python is used to immediately exit the current for or while loop, transferring the program's control to the next line of code after the loop.

```python
#Searching an element in a list
nums = [10,20,30,40,50]
key = int(input("Enter element to search: "))

i = 0
while i < len(nums):
  if nums[i] == key:
    print("Element found at index",i)
    break
  i += 1
else:
  print("Element not found")
```

# **continue**

The continue statement in Python is used to skip the rest of the code inside the current loop iteration and immediately jump to the next iteration.

```python
# Write a python program using while loop to print only odd numbers from 1 to 10 using continue statement

num = 1
while num <= 10:
  if num % 2 == 0: # If the number is even
    num += 1     # Increment to check the next number
    continue     # Skip the rest of the current iteration
  print(num)
  num += 1
```
